## Matplotlib

**Matplotlib**

Matplotlib is a package used to create static, dynamic, and interactive plots.

**pyplot**

Figures can be created using the pyplot library, a state-based interface to the Matplotlib package that uses a syntax similar to MATLAB.

**plot()**

The most common object is a line plot, created using the plot() function (Ex: plt.plot(x, y)), which connects consecutive x- and y- coordinates (x and y must be arrays of the same size).

**scatter()**

Another common object is a scatter plot, created using the scatter() function (Ex: plt.scatter(x, y)), which shows all pairs of x- and y- coordinates.

**figure()**

Figure(): Creates a new figure for a plot to appear in. Ex: plt.figure(figsize=[4,5]).

**show()**

Show(): Displays the figure and all the objects the figure contains. Ex: plt.show().

**savefig(fname)**

Savefig(fname): Saves the figure in the current working directory with the filename fname. Ex: plt.savefig(line_plot.png).

#### Creating a line plot: Adjust figure size

```python
# Load matplotlib.pyplot, pandas, and numpy
import matplotlib.pyplot as plt
import numpy as np

# Create array for x- and y- coordinates
x = [1, 2, 3, 4]
y = [5, 1, 4, 2]

# Create 5x6 figure
plt.figure(figsize=[5,6])

# Create a line plot
plt.plot(x, y)
plt.savefig('plot1.png')
```

#### Creating a line plot: Change plot equation.

```python
# Load matplotlib.pyplot, pandas, and numpy
import matplotlib.pyplot as plt
import numpy as np

# Create an equally spaced array with 100
# elements from -5 to 5
x = np.linspace(-5, 5, num=100)

# Plot the graph for y = x^2
plt.plot(x, x * x)

# Plot the graph for y = -4x - 4
plt.plot(x, -4 * x - 4)
plt.savefig('plot2.png')
```

**title()**

The pyplot title() function adds a title to a figure. Ex: plt.title('Alcohol-related fatalities on highways').

**xlabel() / ylabel()**

The xlabel() and ylabel() functions add x- and y- axis labels to a plot. Ex: plt.xlabel('Year').

**text(x, y, s)**

Text(x, y, s): Adds string s to the figure at coordinates (x, y).

**annotate(s, xy, xytext)**

Annotate(s, xy, xytext): Links string s at coordinates given by xytext to a point given by xy.

**legend(loc)**

Legend(loc): Adds a legend in the figure in a specified location (loc), which can include 'best', 'center', 'upper center', 'lower center', 'center left', 'center right', 'upper left', 'upper right', 'lower left', and 'lower right'. If no loc is given, legend is placed at the 'best' location with the least plot overlap.

```python
df = pd.read_csv('fatalities.csv')
plt.plot(df['Year'], df['Total'], label='Total')
plt.plot(df['Year'], df['Alcohol-related'], label='Alcohol-related')
plt.title('Alcohol-related fatalities on highways', fontsize=20)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Highway fatalities', fontsize=14)
plt.annotate(
    'Drinking age changed to 21',
    arrowprops=dict(),
    xy=(1984, 24762),
    xytext=(1986, 30000))
plt.legend()
```

#### grid()
The pyplot grid() function adds grid lines to a plot.

```python
# Load packages
import matplotlib.pyplot as plt
import pandas as pd

# Create dashed green grid lines
plt.grid(linestyle='--', color='green')

# Load oldfaithfulCluster.csv data
df = pd.read_csv('oldfaithfulCluster.csv')
# Define groups according to clusters
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
# Create scatter plots
plt.scatter(group1['Eruption'], group1['Waiting'], label='1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], label='0', edgecolors='white')
# Add axes labels and title
plt.xlabel('Eruption time', fontsize=14)
plt.ylabel('Waiting time', fontsize=14)
plt.title('Eruption time vs. waiting time', fontsize=20)
# Add legend
plt.legend(title='Cluster')

plt.savefig('eruption_scatter_grid_lines.png')
```

**subplot()**

Multiple plots can be drawn in the same figure using the pyplot subplot() function.

**suptitle()**

The pyplot function suptitle() adds a title to the entire figure, not just the individual subplots.

```python
# Load packages
import matplotlib.pyplot as plt
import pandas as pd

# Load oldfaithfulCluster.csv data
df = pd.read_csv('oldfaithfulCluster.csv')

plt.suptitle('Eruption time vs. waiting time', fontsize=20, c='black')

# Subplot 1
plt.subplot(1, 2, 1)
plt.scatter(df['Eruption'], df['Waiting'])
plt.title('Not clustered', fontsize=12)
plt.xlabel('Eruption time', fontsize=10)
plt.ylabel('Waiting time', fontsize=10)

# Subplot 2
plt.subplot(1, 2, 2)
group1 = df[df['Cluster'] == 1]
group0 = df[df['Cluster'] == 0]
plt.scatter(group1['Eruption'], group1['Waiting'], label='1', edgecolors='white')
plt.scatter(group0['Eruption'], group0['Waiting'], label='0', edgecolors='white')
plt.title('Clustered', fontsize=12)
plt.xlabel('Eruption time', fontsize=10)
plt.legend()

plt.savefig('eruption_scatter_subplots.png')
```
